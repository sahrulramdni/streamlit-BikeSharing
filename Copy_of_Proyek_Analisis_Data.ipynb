## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".


*   **Pertanyaan 1: Optimalisasi Operasional Berdasarkan Jam Sibuk**

***"Jam berapa saja yang termasuk dalam kategori peak hour (puncak tertinggi) yang menyumbang 70% dari total peminjaman sepeda pada hari kerja (workingday = 1) selama periode 2 tahun (2011-2012), dan bagaimana strategi penjadwalan ulang sepeda (rebalancing) yang paling tepat untuk mengatasi lonjakan permintaan di jam-jam tersebut?"***

Keterangan:

- **Specific**: Analisis ini fokus pada hari kerja (workingday = 1) dan ingin mengidentifikasi jam-jam spesifik (misalnya: jam 7-8 pagi, jam 17-18 malam) yang menjadi penyumbang terbesar peminjaman. Bukan sekadar "jam sibuk", tapi jam sibuk pada hari kerja.

- **Measurable**: Targetnya jelas secara kuantitatif, yaitu "70% dari total peminjaman". Tim analis dapat menghitung kumulatif persentase peminjaman per jam dan menentukan batas jam mana saja yang mencapai angka 70%. Metrik yang digunakan adalah cnt (jumlah peminjaman) per hr.

- **Action-Oriented**: Jika peak hour terjadi pada jam 07.00 - 09.00, maka aksi nyata yang bisa dilakukan adalah memastikan truk operasional sudah selesai memindahkan sepeda dari stasiun pusat kota ke stasiun pemukiman sebelum jam 06.30 pagi. Jika puncak sore terjadi pada jam 17.00 - 19.00, maka aksi nyata adalah mempersiapkan tim untuk memindahkan sepeda kembali ke pusat kota pada jam 15.00 - 16.00.

- **Relevant**: Masalah ketidakseimbangan ketersediaan sepeda (penumpukan di satu stasiun dan kekosongan di stasiun lain) adalah inti dari tantangan operasional bisnis bike sharing. Pemecahan masalah ini berdampak langsung pada retensi pelanggan dan efisiensi biaya operasional.

- **Time-bound**: Analisis menggunakan data selama 2 tahun (2011-2012). Rentang waktu ini cukup panjang untuk melihat pola yang konsisten, menghilangkan anomali musiman, dan menghasilkan rekomendasi yang valid untuk jangka panjang.




*   **Pertanyaan 2: Strategi Mitigasi Risiko dan Promosi Berbasis Cuaca**

***"Faktor cuaca spesifik apa (dari kolom weathersit dan hum) yang paling dominan menyebabkan penurunan jumlah peminjaman sepeda lebih dari 50% pada akhir pekan (workingday = 0) di musim gugur (season = 3), dan jenis promo atau insentif apa yang paling efektif untuk mendorong peminjaman di kondisi tersebut?"***

Keterangan:

- **Specific**: Pertanyaan ini tidak sekadar menanyakan "apakah cuaca berpengaruh?". Pertanyaan ini mengerucut pada kondisi spesifik: yaitu weekend (workingday = 0) di musim gugur (season = 3). Ini penting karena persepsi orang terhadap cuaca buruk di akhir pekan (leisure time) berbeda dengan di hari kerja (commute time).

- **Measurable**: Ada angka yang jelas, yaitu "penurunan lebih dari 50%" (dibandingkan dengan rata-rata peminjaman di musim gugur saat cuaca cerah). Analis dapat menghitung rata-rata cnt untuk setiap nilai weathersit (1:Cerah, 2:Berkabut, 3:Hujan Ringan) dan mengukur persentase penurunannya. Kelembaban (hum) juga bisa dianalisis secara numerik.

- **Action-Oriented**: Jika ternyata faktor dominannya adalah "Kelembaban > 80% + Kabut", maka tim marketing bisa mengotomatisasi sistem notifikasi push: "Hey, it's foggy but fun! 20% off for your next ride." Jika faktor dominannya adalah "Hujan Ringan", tim bisa menyiapkan promo "Rainy Day Loyalty Points" (poin ganda) untuk mendorong peminjaman.

- **Relevant**: Data historis menunjukkan penjualan sangat fluktuatif tergantung cuaca. Bagi manajemen, memiliki strategi anti-fragile terhadap cuaca buruk adalah keunggulan kompetitif. Hal ini relevan untuk menjaga pendapatan tetap stabil sepanjang tahun meskipun cuaca sedang tidak bersahabat.

- **Time-bound**: Analisis difokuskan pada musim gugur (season = 3). Musim ini unik karena sering terjadi perubahan cuaca ekstrem (dari cerah ke hujan) yang sulit diprediksi. Dengan menganalisis musim ini secara spesifik, perusahaan dapat mempersiapkan strategi khusus di bulan-bulan September, Oktober, dan November.

```python
# ---------------------------------------------------------------------
# PROBLEM 1: CEK TIPE DATA
# ---------------------------------------------------------------------
print("\n PROBLEM 1 CEK TIPE DATA")
print("-"*50)

print("\nTipe data hour.csv:")
print(df_hour.dtypes)

print("\nTipe data day.csv:")
print(df_day.dtypes)

# Identifikasi kolom dengan tipe data kurang tepat
wrong_types_hour = df_hour.dtypes[df_hour.dtypes == 'object'].index.tolist()
wrong_types_day = df_day.dtypes[df_day.dtypes == 'object'].index.tolist()

print(f"\n Kolom dengan tipe 'object' di hour.csv: {wrong_types_hour}")
print(f" Kolom dengan tipe 'object' di day.csv: {wrong_types_day}")
print("→ Masalah: Kolom tanggal (dteday) seharusnya bertipe datetime, bukan object/string")

# ---------------------------------------------------------------------
# PROBLEM 2: CEK MISSING VALUES
# ---------------------------------------------------------------------
print("\n PROBLEM 2 CEK MISSING VALUES")
print("-"*50)

missing_hour = df_hour.isnull().sum()
missing_day = df_day.isnull().sum()

print(f"\nMissing values di hour.csv: {missing_hour.sum()} dari {len(df_hour)} baris")
print(f"Missing values per kolom di hour.csv:\n{missing_hour[missing_hour > 0]}")

print(f"\nMissing values di day.csv: {missing_day.sum()} dari {len(df_day)} baris")
print(f"Missing values per kolom di day.csv:\n{missing_day[missing_day > 0]}")

if missing_hour.sum() == 0 and missing_day.sum() == 0:
    print("\n Tidak ditemukan missing values pada kedua dataset!")

# ---------------------------------------------------------------------
# PROBLEM 3: CEK DATA DUPLIKAT
# ---------------------------------------------------------------------
print("\n PROBLEM 3 CEK DATA DUPLIKAT")
print("-"*50)

duplicate_hour = df_hour.duplicated().sum()
duplicate_day = df_day.duplicated().sum()

print(f"Data duplikat di hour.csv: {duplicate_hour} baris")
print(f"Data duplikat di day.csv: {duplicate_day} baris")

if duplicate_hour == 0 and duplicate_day == 0:
    print("\n Tidak ditemukan data duplikat pada kedua dataset")

# ---------------------------------------------------------------------
# PROBLEM 4: CEK OUTLIER DENGAN METODE IQR
# ---------------------------------------------------------------------
print("\n PROBLEM 4 CEK OUTLIER PADA KOLOM NUMERIK")
print("-"*50)

def check_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return len(outliers), len(outliers)/len(data)*100, lower_bound, upper_bound

numeric_columns = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

print("\nHasil deteksi outlier pada hour.csv:")
print("-"*40)
for col in numeric_columns:
    n_outliers, pct_outliers, lb, ub = check_outliers(df_hour, col)
    status = " ADA OUTLIER" if n_outliers > 0 else " AMAN"
    print(f"{col:12} | Outlier: {n_outliers:5} ({pct_outliers:5.2f}%) | {status}")

# ---------------------------------------------------------------------
# PROBLEM 5: CEK KONSISTENSI DATA (casual + registered = cnt)
# ---------------------------------------------------------------------
print("\n PROBLEM 5 CEK KONSISTENSI DATA")
print("-"*50)

inconsistent = df_hour[df_hour['casual'] + df_hour['registered'] != df_hour['cnt']]
print(f"Jumlah baris dengan casual + registered ≠ cnt: {len(inconsistent)}")

if len(inconsistent) == 0:
    print(" Data konsisten, casual + registered = cnt untuk semua baris.")
else:
    print(f" Ditemukan {len(inconsistent)} baris yang tidak konsisten!")
    display(inconsistent.head())

# ---------------------------------------------------------------------
# PROBLEM 6: CEK KONSISTENSI HOLIDAY VS WORKINGDAY
# ---------------------------------------------------------------------
print("\n PROBLEM 6 CEK KONSISTENSI HOLIDAY DAN WORKINGDAY")
print("-"*50)

# Holiday seharusnya tidak termasuk workingday
inconsistent_holiday = df_hour[(df_hour['holiday'] == 1) & (df_hour['workingday'] == 1)]
print(f"Jumlah baris dengan holiday=1 dan workingday=1: {len(inconsistent_holiday)}")

if len(inconsistent_holiday) == 0:
    print(" Konsisten, Holiday tidak termasuk workingday.")
else:
    print(" Ada inkonsistensi: holiday seharusnya bukan hari kerja")

# ---------------------------------------------------------------------
# PROBLEM 7: CEK RENTANG NILAI KOLOM (VALIDASI)
# ---------------------------------------------------------------------
print("\n PROBLEM 7 CEK RENTANG NILAI KOLOM")
print("-"*50)

# Cek rentang untuk kolom yang seharusnya antara 0-1 (sudah dinormalisasi)
normalized_cols = ['temp', 'atemp', 'hum', 'windspeed']

for col in normalized_cols:
    min_val = df_hour[col].min()
    max_val = df_hour[col].max()
    is_valid = (min_val >= 0) and (max_val <= 1)
    status = "✅" if is_valid else "⚠️"
    print(f"{status} {col:12} | min: {min_val:.3f} | max: {max_val:.3f} | range: 0-1")

# ---------------------------------------------------------------------
# PROBLEM 8: CEK NILAI UNIK PADA KOLOM KATEGORIKAL
# ---------------------------------------------------------------------
print("\n PROBLEM 8 CEK NILAI UNIK KOLOM KATEGORIKAL")
print("-"*50)

categorical_cols = ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']
expected_values = {
    'season': [1, 2, 3, 4],
    'yr': [0, 1],
    'mnth': list(range(1, 13)),
    'hr': list(range(0, 24)),
    'holiday': [0, 1],
    'weekday': list(range(0, 7)),
    'workingday': [0, 1],
    'weathersit': [1, 2, 3, 4]
}

for col in categorical_cols:
    unique_vals = sorted(df_hour[col].unique())
    expected = expected_values[col]
    is_valid = unique_vals == expected
    status = "✅" if is_valid else "⚠️"
    print(f"{status} {col:12} | unique: {unique_vals}")

# ---------------------------------------------------------------------
# RINGKASAN MASALAH
# ---------------------------------------------------------------------
print("\n" + "="*70)
print("RINGKASAN MASALAH YANG DITEMUKAN")
print("="*70)

problems_summary = pd.DataFrame([
    {"No": 1, "Masalah": "Tipe data 'dteday' masih object", "Dampak": "Tidak bisa operasi datetime"},
    {"No": 2, "Masalah": "8 kolom kategorikal bertipe int64", "Dampak": "Boros memori, query lambat"},
    {"No": 3, "Masalah": f"Outlier pada kolom 'cnt': {check_outliers(df_hour, 'cnt')[0]} points", "Dampak": "Mempengaruhi statistik mean/std"},
    {"No": 4, "Masalah": "Tidak ada missing values", "Dampak": "-"},
    {"No": 5, "Masalah": "Tidak ada data duplikat", "Dampak": "-"},
    {"No": 6, "Masalah": "Data konsisten (casual+registered=cnt)", "Dampak": "-"},
])

display(problems_summary)
```

```python
# menggunakan data yang sudah dibersihkan
df = df_hour_clean.copy()

# Konversi kolom kategorikal ke numeric untuk keperluan filtering
for col in ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']:
    if col in df.columns:
        if df[col].dtype.name == 'category':
            df[col] = df[col].astype(int)

print(f"   Jumlah baris: {len(df):,}")
print(f"   Jumlah kolom: {len(df.columns)}")
print(f"   Rentang waktu: {df['dteday'].min()} s.d {df['dteday'].max()}")

# ============================================
# 1 EDA UNTUK PERTANYAAN BISNIS 1: PEAK HOUR PADA HARI KERJA
# ============================================

print("\n" + "="*70)
print("1 EDA - PERTANYAAN BISNIS 1: IDENTIFIKASI PEAK HOUR")
print("="*70)

print("\n Pertanyaan Bisnis 1")
print("Jam berapa saja yang termasuk dalam kategori peak hour yang menyumbang")
print("70% dari total peminjaman sepeda pada hari kerja (workingday=1) selama 2 tahun?")

# Filter data untuk hari kerja saja (workingday = 1)
workingday_df = df[df['workingday'] == 1]
print(f"\n Data hari kerja: {len(workingday_df):,} baris")

# Hitung total peminjaman per jam pada hari kerja
hourly_total = workingday_df.groupby('hr')['cnt'].sum().reset_index()
hourly_total.columns = ['hour', 'total_cnt']

# Hitung persentase kontribusi per jam
hourly_total['percentage'] = (hourly_total['total_cnt'] / hourly_total['total_cnt'].sum()) * 100

# Hitung persentase kumulatif (dari jam 0 sampai 23)
hourly_total['cumulative_pct'] = hourly_total['percentage'].cumsum()

# Identifikasi jam yang termasuk peak hour (sampai cumulative <= 70)
hourly_total['is_peak'] = hourly_total['cumulative_pct'] <= 70

# Tampilkan hasil
print("\n TOTAL PEMINJAMAN PER JAM PADA HARI KERJA:")
print("-"*80)
print(f"{'Jam':<6} {'Total Peminjaman':<18} {'Kontribusi':<12} {'Kumulatif':<12} {'Status':<10}")
print("-"*80)

for _, row in hourly_total.iterrows():
    status = "PEAK HOUR" if row['is_peak'] else "-"
    print(f"{row['hour']:<6} {row['total_cnt']:>15,}  {row['percentage']:>10.2f}%  {row['cumulative_pct']:>10.2f}%  {status:<10}")

# Dapatkan daftar peak hours
peak_hours_list = hourly_total[hourly_total['is_peak']]['hour'].tolist()
peak_contribution = hourly_total[hourly_total['is_peak']]['percentage'].sum()

print("\n" + "="*60)
print(" KESIMPULAN SEMENTARA PERTANYAAN 1:")
print(f"   Peak hours (menyumbang {peak_contribution:.1f}% peminjaman): {peak_hours_list}")
print(f"   Jumlah peak hours: {len(peak_hours_list)} jam")
print(f"   Jam dengan peminjaman tertinggi: Jam 08.00 dan Jam 17.00-18.00")
print("="*60)

# ============================================
# 2 EDA UNTUK PERTANYAAN BISNIS 2: DAMPAK CUACA
# ============================================

print("\n" + "="*70)
print("2 EDA - PERTANYAAN BISNIS 2: DAMPAK CUACA PADA AKHIR PEKAN DI MUSIM GUGUR")
print("="*70)

print("\n Pertanyaan Bisnis 2")
print("Faktor cuaca spesifik apa yang paling dominan menyebabkan penurunan")
print("jumlah peminjaman sepeda lebih dari 50% pada akhir pekan di musim gugur?")

# Filter data: akhir pekan (workingday=0) DAN musim gugur (season=3)
weekend_fall_df = df[(df['workingday'] == 0) & (df['season'] == 3)]

print(f"\n DATA YANG DIANALISIS:")
print(f"   - Filter: workingday = 0 (Akhir Pekan/Libur)")
print(f"   - Filter: season = 3 (Musim Gugur)")
print(f"   - Jumlah data points: {len(weekend_fall_df):,} jam")
print(f"   - Rata-rata peminjaman: {weekend_fall_df['cnt'].mean():.1f} sepeda/jam")

# Analisis per kondisi cuaca
weather_analysis = weekend_fall_df.groupby('weathersit').agg({
    'cnt': ['mean', 'std', 'min', 'max', 'count'],
    'hum': ['mean', 'min', 'max'],
    'temp': ['mean', 'min', 'max']
}).round(2)

weather_analysis.columns = ['mean_cnt', 'std_cnt', 'min_cnt', 'max_cnt', 'count',
                            'mean_hum', 'min_hum', 'max_hum',
                            'mean_temp', 'min_temp', 'max_temp']
weather_analysis = weather_analysis.reset_index()

# Mapping weathersit ke label
weather_labels = {1: 'Cerah/Berawan', 2: 'Berkabut', 3: 'Hujan Ringan', 4: 'Hujan Lebat'}
weather_analysis['weather_label'] = weather_analysis['weathersit'].map(weather_labels)

# Hitung persentase penurunan terhadap cuaca cerah (weathersit=1)
clear_mean = weather_analysis[weather_analysis['weathersit'] == 1]['mean_cnt'].values[0]
weather_analysis['decline_pct'] = ((clear_mean - weather_analysis['mean_cnt']) / clear_mean) * 100
weather_analysis['decline_pct'] = weather_analysis['decline_pct'].clip(lower=0)

print("\n ANALISIS PER KONDISI CUACA:")
print("-"*100)
print(f"{'Kondisi Cuaca':<18} {'Rata-rata':<12} {'Std':<10} {'Min':<8} {'Max':<8} {'Penurunan':<12}")
print("-"*100)

for _, row in weather_analysis.iterrows():
    decline_status = "⚠️ >50%" if row['decline_pct'] > 50 else "✅"
    print(f"{row['weather_label']:<18} {row['mean_cnt']:>8.0f} sep   {row['std_cnt']:>8.2f}   {row['min_cnt']:>6.0f}   {row['max_cnt']:>6.0f}   {row['decline_pct']:>5.1f}% {decline_status}")

print("\n ANALISIS FAKTOR KELEMBABAN PER KONDISI CUACA:")
print("-"*70)
print(f"{'Kondisi Cuaca':<18} {'Rata-rata Kelembaban':<22} {'Range Kelembaban':<20}")
print("-"*70)

for _, row in weather_analysis.iterrows():
    print(f"{row['weather_label']:<18} {row['mean_hum']:>10.3f}         {row['min_hum']:.2f} - {row['max_hum']:.2f}")

print("\n ANALISIS FAKTOR SUHU PER KONDISI CUACA:")
print("-"*70)
print(f"{'Kondisi Cuaca':<18} {'Rata-rata Suhu':<18} {'Range Suhu':<18}")
print("-"*70)

for _, row in weather_analysis.iterrows():
    print(f"{row['weather_label']:<18} {row['mean_temp']:>10.3f}         {row['min_temp']:.2f} - {row['max_temp']:.2f}")

# Identifikasi faktor dominan
print("\n" + "="*60)
print(" KESIMPULAN SEMENTARA PERTANYAAN 2:")
print("   Faktor dominan penyebab penurunan >50%:")

dominant_factors = []
for _, row in weather_analysis.iterrows():
    if row['decline_pct'] > 50:
        print(f"   - {row['weather_label']}: Penurunan {row['decline_pct']:.1f}%")
        print(f"     (Kelembaban rata-rata: {row['mean_hum']:.2f}, Suhu rata-rata: {row['mean_temp']:.2f})")
        dominant_factors.append(row['weather_label'])

if len(dominant_factors) == 0:
    print("   - Tidak ada kondisi cuaca dengan penurunan >50% dalam data ini")

print("="*60)

# ============================================
# 3 ANALISIS TAMBAHAN: RINGKASAN STATISTIK DESKRIPTIF
# ============================================

print("\n" + "="*70)
print("3 ANALISIS TAMBAHAN: RINGKASAN STATISTIK DESKRIPTIF")
print("="*70)

# Statistik deskriptif untuk kolom numerik
print("\n STATISTIK DESKRIPTIF KOLOM NUMERIK:")
print("-"*60)
numeric_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
display(df[numeric_cols].describe())

# Statistik per kelompok (grouping)
print("\n RATA-RATA PEMINJAMAN PER MUSIM:")
print("-"*40)
season_stats = df.groupby('season').agg({
    'cnt': ['mean', 'std', 'count'],
    'temp': 'mean'
}).round(2)
season_stats.columns = ['mean_cnt', 'std_cnt', 'count', 'mean_temp']
season_stats = season_stats.reset_index()
season_stats['season_label'] = season_stats['season'].map({1:'Spring', 2:'Summer', 3:'Fall', 4:'Winter'})
display(season_stats[['season_label', 'mean_cnt', 'std_cnt', 'mean_temp']])

print("\n RATA-RATA PEMINJAMAN PER TIPE HARI:")
print("-"*40)
day_type_stats = df.groupby('workingday').agg({
    'cnt': ['mean', 'std', 'count']
}).round(2)
day_type_stats.columns = ['mean_cnt', 'std_cnt', 'count']
day_type_stats = day_type_stats.reset_index()
day_type_stats['day_type'] = day_type_stats['workingday'].map({1:'Hari Kerja', 0:'Akhir Pekan/Libur'})
display(day_type_stats[['day_type', 'mean_cnt', 'std_cnt']])

print("\n RATA-RATA PEMINJAMAN PER TAHUN:")
print("-"*40)
year_stats = df.groupby('yr').agg({
    'cnt': ['mean', 'std', 'count', 'sum']
}).round(2)
year_stats.columns = ['mean_cnt', 'std_cnt', 'count', 'total_cnt']
year_stats = year_stats.reset_index()
year_stats['year_label'] = year_stats['yr'].map({0:'2011', 1:'2012'})
display(year_stats[['year_label', 'mean_cnt', 'total_cnt']])

# ============================================
# 4 ANALISIS KORELASI (UNTUK IDENTIFIKASI HUBUNGAN)
# ============================================

print("\n" + "="*70)
print("4 ANALISIS KORELASI ANTAR VARIABEL")
print("="*70)

# Hitung korelasi dengan cnt (total peminjaman)
numeric_for_corr = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
corr_with_cnt = df[numeric_for_corr].corr()['cnt'].sort_values(ascending=False)

print("\n KORELASI DENGAN TOTAL PEMINJAMAN (cnt):")
print("-"*50)
for var, corr in corr_with_cnt.items():
    if var != 'cnt':
        strength = "sangat kuat" if abs(corr) > 0.7 else "kuat" if abs(corr) > 0.5 else "sedang" if abs(corr) > 0.3 else "lemah"
        direction = "positif" if corr > 0 else "negatif"
        print(f"   - {var}: {corr:.3f} ({direction} {strength})")

print("\n INTERPRETASI KORELASI:")
print("-"*50)
print("    registered (+0.99) → Hampir semua peminjaman dari pengguna terdaftar")
print("    temp (+0.54) → Suhu hangat meningkatkan peminjaman")
print("    casual (+0.48) → Pengguna kasual cukup berkontribusi")
print("    hum (-0.35) → Kelembaban tinggi menurunkan peminjaman")
print("    windspeed (-0.10) → Angin berpengaruh sangat kecil")

# Hitung nilai untuk summary
count_peak = len(peak_hours_list)
pct_peak = peak_contribution
peak_hours = f"{peak_hours_list[0]}-{peak_hours_list[-1]}"

clear_mean = weather_analysis[weather_analysis['weathersit'] == 1]['mean_cnt'].values[0]

mist_data = weather_analysis[weather_analysis['weathersit'] == 2]
if len(mist_data) > 0:
    mist_decline = mist_data['decline_pct'].values[0]
    mist_hum = mist_data['mean_hum'].values[0]
else:
    mist_decline = 0
    mist_hum = 0

rain_data = weather_analysis[weather_analysis['weathersit'] == 3]
if len(rain_data) > 0:
    rain_decline = rain_data['decline_pct'].values[0]
    rain_hum = rain_data['mean_hum'].values[0]
else:
    rain_decline = 0
    rain_hum = 0

fall_mean = season_stats[season_stats['season'] == 3]['mean_cnt'].values[0]
spring_mean = season_stats[season_stats['season'] == 1]['mean_cnt'].values[0]

total_2011 = year_stats[year_stats['yr'] == 0]['total_cnt'].values[0]
total_2012 = year_stats[year_stats['yr'] == 1]['total_cnt'].values[0]
increase_pct = ((total_2012 - total_2011) / total_2011) * 100
```

```python
# ============================================
# 1 RFM ANALYSIS (UNTUK DATA HARIAN)
# ============================================

print("\n 1 RFM ANALYSIS (Recency, Frequency, Monetary)")
print("-"*50)

# Gunakan data harian (day.csv) untuk analisis RFM
df_day_clean = df_day_clean.copy()
df_day_clean['dteday'] = pd.to_datetime(df_day_clean['dteday'])

# Tentukan tanggal referensi (tanggal terakhir dalam dataset)
reference_date = df_day_clean['dteday'].max()

# Hitung Recency (hari sejak terakhir peminjaman)
rfm = df_day_clean.groupby('dteday').agg({
    'instant': 'count',  # Frequency
    'cnt': 'sum'         # Monetary
}).reset_index()
rfm.columns = ['dteday', 'frequency', 'monetary']

# Hitung recency (hari sejak tanggal referensi)
rfm['recency'] = (reference_date - rfm['dteday']).dt.days

# Buat score (1-4) untuk setiap metrik
rfm['R_score'] = pd.qcut(rfm['recency'], 4, labels=['4', '3', '2', '1']).astype(int)
rfm['F_score'] = pd.qcut(rfm['frequency'].rank(method='first'), 4, labels=['1', '2', '3', '4']).astype(int)
rfm['M_score'] = pd.qcut(rfm['monetary'], 4, labels=['1', '2', '3', '4']).astype(int)

# Hitung total RFM score
rfm['RFM_score'] = rfm['R_score'] + rfm['F_score'] + rfm['M_score']

# Segmentasi pelanggan
def rfm_segmentation(row):
    if row['R_score'] >= 3 and row['F_score'] >= 3 and row['M_score'] >= 3:
        return 'Champions (Paling Aktif)'
    elif row['R_score'] >= 3 and row['F_score'] >= 2:
        return 'Loyal (Setia)'
    elif row['R_score'] <= 2 and row['F_score'] >= 3 and row['M_score'] >= 3:
        return 'At Risk (Berisiko)'
    elif row['R_score'] <= 2 and row['F_score'] <= 2:
        return 'Lost (Hilang)'
    else:
        return 'Others'

rfm['segment'] = rfm.apply(rfm_segmentation, axis=1)

print("\n DISTRIBUSI SEGMEN PELANGGAN:")
segment_counts = rfm['segment'].value_counts()
for segment, count in segment_counts.items():
    pct = (count / len(rfm)) * 100
    print(f"   - {segment}: {count} hari ({pct:.1f}%)")

# Visualisasi RFM
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# R-Score Distribution
axes[0].hist(rfm['R_score'], bins=4, edgecolor='black', color='steelblue', alpha=0.7)
axes[0].set_title('Distribusi Recency Score', fontsize=12)
axes[0].set_xlabel('R-Score (1=terlama, 4=terbaru)')
axes[0].set_ylabel('Jumlah Hari')

# F-Score Distribution
axes[1].hist(rfm['F_score'], bins=4, edgecolor='black', color='coral', alpha=0.7)
axes[1].set_title('Distribusi Frequency Score', fontsize=12)
axes[1].set_xlabel('F-Score (1=terendah, 4=tertinggi)')
axes[1].set_ylabel('Jumlah Hari')

# M-Score Distribution
axes[2].hist(rfm['M_score'], bins=4, edgecolor='black', color='green', alpha=0.7)
axes[2].set_title('Distribusi Monetary Score', fontsize=12)
axes[2].set_xlabel('M-Score (1=terendah, 4=tertinggi)')
axes[2].set_ylabel('Jumlah Hari')

plt.tight_layout()
plt.show()

# ============================================
# 2 CLUSTERING BERDASARKAN JAM DAN CUACA
# ============================================

print("\n 2 CLUSTERING POLA PEMINJAMAN")
print("-"*50)

# Buat pivot table untuk melihat pola peminjaman per jam per kondisi cuaca
hour_weather_pivot = df.groupby(['hr', 'weather_label'])['cnt'].mean().reset_index()
hour_weather_pivot = hour_weather_pivot.pivot(index='hr', columns='weather_label', values='cnt').fillna(0)

print("\n RATA-RATA PEMINJAMAN PER JAM PER KONDISI CUACA:")
print("-"*70)
print(hour_weather_pivot.round(0))

# Identifikasi cluster jam berdasarkan pola peminjaman
def classify_hour(hour):
    if 6 <= hour <= 9:
        return 'Morning Peak (06:00-09:00)'
    elif 17 <= hour <= 19:
        return 'Evening Peak (17:00-19:00)'
    elif 10 <= hour <= 16:
        return 'Midday (10:00-16:00)'
    elif 20 <= hour <= 23:
        return 'Night (20:00-23:00)'
    else:
        return 'Early Morning/Late Night (00:00-05:00)'

df['hour_cluster'] = df['hr'].apply(classify_hour)

# Hitung rata-rata peminjaman per cluster
cluster_stats = df.groupby('hour_cluster').agg({
    'cnt': ['mean', 'std', 'count']
}).round(2)
cluster_stats.columns = ['mean_cnt', 'std_cnt', 'count']
cluster_stats = cluster_stats.reset_index()

print("\n STATISTIK PER CLUSTER JAM:")
print("-"*70)
for _, row in cluster_stats.iterrows():
    print(f"   {row['hour_cluster']}:")
    print(f"      - Rata-rata: {row['mean_cnt']:.0f} sepeda/jam")
    print(f"      - Std Dev: {row['std_cnt']:.0f}")
    print(f"      - Jumlah data: {row['count']:,} jam")

# Visualisasi clustering
fig, ax = plt.subplots(figsize=(12, 6))

cluster_colors = {
    'Morning Peak (06:00-09:00)': '#e74c3c',
    'Evening Peak (17:00-19:00)': '#e67e22',
    'Midday (10:00-16:00)': '#3498db',
    'Night (20:00-23:00)': '#2ecc71',
    'Early Morning/Late Night (00:00-05:00)': '#95a5a6'
}

hourly_mean = df.groupby('hr')['cnt'].mean().reset_index()
hourly_mean['cluster'] = hourly_mean['hr'].apply(classify_hour)

for cluster, color in cluster_colors.items():
    cluster_data = hourly_mean[hourly_mean['cluster'] == cluster]
    ax.scatter(cluster_data['hr'], cluster_data['cnt'],
               color=color, s=100, alpha=0.7, label=cluster, edgecolors='white', linewidth=1.5)

ax.plot(hourly_mean['hr'], hourly_mean['cnt'], color='black', alpha=0.3, linewidth=1)

ax.set_title('Pengelompokan Pola Peminjaman Berdasarkan Jam (Clustering)',
             fontsize=14, fontweight='bold')
ax.set_xlabel('Jam', fontsize=12)
ax.set_ylabel('Rata-rata Jumlah Peminjaman', fontsize=12)
ax.set_xticks(range(0, 24))
ax.set_xticklabels([f'{i:02d}:00' for i in range(0, 24)], rotation=45)
ax.legend(loc='upper left', fontsize=9)
ax.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()
```
